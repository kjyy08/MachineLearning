```python
import tensorflow as tf

# tf.constant() == C언어 const와 같음
# tf.variable() == C언어 일반 변수 선언과 같음
hello = tf.constant("Hello world!")
print(hello)

print(hello.numpy())

print(hello.numpy().decode('utf-8'))
```

```python
# range(시작값, 종료값, 증감값)
for i in range(5):
  print(i, end='')
print('\n')

for i in range(3, 5):
  print(i, end='')
print("\n")

for i in range(3, 5, 2):
  print(i, end='')
```

```python
i = 0
while i < 5:
  print(i, end='')
  i += 1
print('\n')

i = 3
while i < 5:
  print(i, end='')
  i += 1;
print("\n")

i = 3
while i < 5:
  print(i, end='')
  i += 2
```

```python
score = int(input("점수 입력: "))

if score >= 90:
  print("A", end='')
else:
  if score >= 80:
    print("B", end='')
  else:
    if score >= 70:
      print("C", end='')
    else:
      if score >= 60:
        print("D", end='')
      else:
        print("F", end='')

print("학점입니다")
```
