```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

x_data = np.array([1.0, 2.0, 3.0], dtype=np.float32)
y_data = np.array([1.0, 2.0, 3.0], dtype=np.float32)
# W = tf.Variable(tf.random.normal([1]))

W_val = []
cost_val = []

for i in range(-30, 50):
  W_curr = i * 0.1
  # tf.square => 제곱 함수
  cost_curr = tf.reduce_mean(tf.square(W_curr * x_data - y_data))

  W_val.append(W_curr)
  cost_val.append(cost_curr)

plt.plot(W_val, cost_val)
plt.xlabel('W')
plt.ylabel('cost')
plt.show()
```

```python
# 텐서플로우 안에 있는 함수 활용

import tensorflow as tf
import matplotlib.pyplot as plt

# 모델을 y = Wx + b로 추정
x_data = [1, 2, 3, 4]
y_data = [2, 1, 4, 3]

# W, b에 임의의 수로 정규분포를 가지는 랜덤 변수를 초기값으로 지정
W = tf.Variable(tf.random.normal([1]))
b = tf.Variable(tf.random.normal([1]))
learning_rate = 0.01

# 경사하강 알고리즘 함수
def gradientDescent():
  with tf.GradientTape() as tape:
    model = W * x_data + b;

    cost = tf.reduce_mean(tf.square(model - y_data))

    gradients = tape.gradient(cost, [W, b])

    tf.optimizers.SGD(learning_rate).apply_gradients(zip(gradients, [W, b]))

for step in range(1501):

  gradientDescent()
  model = W * x_data + b
  cost = tf.reduce_mean(tf.square(model - y_data))

  if step % 50 == 0:
    print(step, W.numpy(), cost.numpy())

plt.scatter(x_data, y_data)
plt.plot(x_data, model, color='green')
plt.show()
```

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np

x_data = [[73, 80, 75], [93, 88, 93], [89, 91, 90], [96, 98, 100], [73, 66, 70]]
y_data = [[152, ], [185, ], [180, ], [196, ], [142, ]]
x_data = np.array(x_data, dtype=np.float32)
y_data = np.array(y_data, dtype=np.float32)

W = tf.Variable(tf.random.normal([3, 1]))
b = tf.Variable(tf.random.normal([1]))
learning_rate = 1e-5

def gradientDescent():
  with tf.GradientTape() as tape:
    # model: H = Wx + b
    model = tf.matmul(x_data, W) + b
    cost = tf.reduce_mean(tf.square(model - y_data))

    gradients = tape.gradient(cost, [W, b])

    tf.optimizers.SGD(learning_rate).apply_gradients(zip(gradients, [W, b]))

cost_vals = []
steps = []

for step in range(10001):
  gradientDescent()
  cost_vals.append(cost)
  steps.append(step)

  if step % 2000 == 0:
    model = tf.matmul(x_data, W) + b
    cost = tf.reduce_mean(tf.square(model - y_data))

    print(step, '\nW=', W.numpy(), "\nmodel: ", model.numpy(), '\nb=', b.numpy(), '\ncost=', cost.numpy(), end="\n-----------------------------------------\n")


plt.plot(W, cost_vals, color='green')
plt.show()
```
