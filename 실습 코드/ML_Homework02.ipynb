```python
# problem 1. 비만 여부를 판단하는 모델 작성
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np

xy = np.loadtxt('data-03-diabetes.csv', delimiter=',', dtype=np.float32)
x_data = xy[:, 0:-1]
y_data = xy[:, [-1]]
x = np.array(x_data, dtype=np.float32)
y = np.array(y_data, dtype=np.float32)

W = tf.Variable(tf.random.normal([8, 1]))
b = tf.Variable(tf.random.normal([1]))
learning_rate = 0.01

def logisticRegression():
  with tf.GradientTape() as tape:
    hyp = tf.sigmoid(tf.matmul(x, W) + b)
    cost = -tf.reduce_mean(y * tf.math.log(hyp) + (1 - y) * tf.math.log(1 - hyp))
    gradients = tape.gradient(cost, [W, b])
    tf.optimizers.SGD(learning_rate).apply_gradients(zip(gradients, [W, b]))

step_list = []
accuracy_list = []
cost_list = []

for step in range(50001):
  logisticRegression()
  hyp = tf.sigmoid(tf.matmul(x, W) + b)
  cost = -tf.reduce_mean(y * tf.math.log(hyp) + (1 - y) * tf.math.log(1 - hyp))
  prediction = tf.cast(hyp > 0.5, dtype=tf.float32) # tf.cast(): 0.5보다 작으면 0, 크면 1을 반환
  accuracy = tf.reduce_mean(tf.cast(tf.equal(prediction, y), dtype= np.float32))

  if step % 1000 == 0:
    step_list.append(step)
    accuracy_list.append(accuracy)
    cost_list.append(cost)

    print("step: ", step, "\nAccuracy: ", accuracy.numpy(), "\nCost: ", cost.numpy(), end="\n---------------------------------\n")

# accuracy
plt.plot(step_list, accuracy_list)
plt.xlabel('step')
plt.ylabel('accuracy')
plt.show()
# cost
plt.plot(step_list, cost_list)
plt.xlabel('step')
plt.ylabel('cost')
plt.show()

```

```python
# problem 2. 동물을 분류하는 모델 작성
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np

xy = np.loadtxt('data-04-zoo.csv', delimiter=',', dtype=np.float32)
x_data = xy[:, 0:-1]
y_data = xy[:, [-1]]
x = np.array(x_data, dtype=np.float32)

animal_type = 7
one_hot_y_data = tf.one_hot(y_data, animal_type)
one_hot_y = tf.reshape(one_hot_y_data, [-1, animal_type])

W = tf.Variable(tf.random.normal([16, 7]))
b = tf.Variable(tf.random.normal([7]))

learning_rate = 0.01

def softmax_classfier():
  with tf.GradientTape() as tape:
    model_LC = tf.matmul(x, W) + b
    cost = tf.reduce_sum(tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=model_LC, labels=one_hot_y)))
    gradients = tape.gradient(cost,(W,b))
    tf.optimizers.SGD(learning_rate).apply_gradients(zip(gradients,(W,b)))

step_list = []
accuracy_list = []
cost_list = []

for step in range(20001):
  softmax_classfier()
  model_LC = tf.matmul(x, W) + b
  model = tf.argmax(tf.nn.softmax(model_LC), 1)
  cost = tf.reduce_sum(tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=model_LC, labels=one_hot_y)))
  accuracy = tf.reduce_mean(tf.cast(tf.equal(model, tf.argmax(one_hot_y, 1)), tf.float32))

  if step % 1000 == 0:
    step_list.append(step)
    accuracy_list.append(accuracy)
    cost_list.append(cost)

    print("step : ", step, "\nmodel: ", model.numpy(), "\naccuracy: ", accuracy.numpy(), "\ncost: ", cost.numpy(), end="\n-----------------------------------------------------------------\n")

    for p, y in zip(model, y_data.flatten()):
      print("[{}] prediction: {}, True Y: {}".format(p==int(y), p, int(y)))

# accuracy
plt.plot(step_list, accuracy_list)
plt.xlabel('step')
plt.ylabel('accuracy')
plt.show()
# cost
plt.plot(step_list, cost_list)
plt.xlabel('step')
plt.ylabel('cost')
plt.show()
```
