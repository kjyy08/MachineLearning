```python
import matplotlib as plt
import numpy as np

plt.rcParams['figure.figsize'] = (12, 6)

# x range = -2.0 ~ 2.0
x = [0.01 * i for i in range(-200, 201)]

# y range = -2.0 ~ 2.0에 대한 sin(x)값
y = [np.sin(j * np.pi) for j in x]

plt.pyplot.plot(x, y)
plt.pyplot.xlabel('x')
plt.pyplot.ylabel('sin(x)')
plt.pyplot.show()
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

x_data = np.array([1.0, 2.0, 3.0], dtype=np.float32)
y_data = np.array([1.0, 2.0, 3.0], dtype=np.float32)
# W = tf.Variable(tf.random.normal([1]))

W_val = []
cost_val = []

for i in range(-30, 50):
  W_curr = i * 0.1
  # tf.squrre => 제곱 함수
  cost_curr = tf.reduce_mean(tf.square(W_curr * x_data - y_data))

  W_val.append(W_curr)
  cost_val.append(cost_curr)

plt.plot(W_val, cost_val)
plt.xlabel('W')
plt.ylabel('cost')
plt.show()
```
